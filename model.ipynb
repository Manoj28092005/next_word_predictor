```python
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
import pandas as pd
data=gutenberg.raw('shakespeare-hamlet.txt')
print(data)
with open('hamlet.txt','w') as f:
    f.write(data)
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
with open('hamlet.txt')as file:
  text=file.read().lower()
tokenizer=Tokenizer()
tokenizer.fit_on_texts([text])
print(tokenizer.word_index)
total_tokens=len(tokenizer.word_index)
print(total_tokens)
```

```python
text=text.split('\n')
```

```python
input=[]
for i in text:
  token_list=tokenizer.texts_to_sequences([i])[0]
  for j in range(1,len(token_list)):
    input.append(token_list[:j+1])
```

```python
input
```

```python
max_len=max([len(i) for i in input])
max_len
```

```python
input=np.array(pad_sequences(input,maxlen=max_len,padding='pre'))
input
```

```python
X,y=input[:,:-1],input[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,Dropout
```

```python
model=Sequential()
model.add(Embedding(total_tokens+1,100,input_length=max_len-1))
model.add(LSTM(150,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dense(total_tokens+1,activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
import tensorflow as tf
with tf.device('/GPU:0'):
  hist=model.fit(x_train,y_train,epochs=100,validation_data=(x_test,y_test))
```

```python
model.save('my_model.h5')
```
